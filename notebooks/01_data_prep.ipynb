```python
import os 
import pandas as pd
```

```python
raw_dir = "../data/raw"
processed_dir = "../data/processed"
```

```python
consumption_list = []
years = range(2018,2026)
```

```python
for year in years:
    consumption_path = os.path.join(raw_dir, f"consumption_data_{year}.csv")
    if os.path.exists(consumption_path):
        df = pd.read_csv(consumption_path)
        consumption_list.append(df)
    else:
        print(f"File not found: {consumption_path}. Check the file path.")
```

```python
consumption_df = pd.concat(consumption_list, ignore_index=True)
```

```python
consumption_df.info()
```

```python
consumption_df.isna().sum()
```

```python
consumption_df.duplicated().sum()
```

```python
consumption_df.head()
```

```python
consumption_df.tail()
```

```python
generation_list = []

for year in years:
    generation_path = os.path.join(raw_dir, f"generation_data_{year}.csv")
    if os.path.exists(generation_path):
        df = pd.read_csv(generation_path)
        generation_list.append(df)
    else:
        print(f"File not found: {generation_path}. Check the file path.")
```

```python
generation_df = pd.concat(generation_list, ignore_index=True)
```

```python
generation_df.info()
```

```python
generation_df.duplicated().sum()
```

```python
generation_df.isna().sum()
```

```python
generation_df = generation_df.rename(columns={"total":"total_generation"})
```

```python
generation_df.head()
```

```python
generation_df.tail()
```

```python
consumption_df["date"] = pd.to_datetime(consumption_df["date"], format="%Y-%m-%dT%H:%M:%S+03:00")
```

```python
print(consumption_df['date'].dtype)
```

```python
consumption_df = consumption_df.sort_values("date").reset_index(drop=True)
```

```python
print(consumption_df["date"].min())
print(consumption_df["date"].max())
```

```python
generation_df["date"] = pd.to_datetime(generation_df["date"], format="%Y-%m-%dT%H:%M:%S+03:00")
```

```python
print(generation_df['date'].dtype)
```

```python
generation_df = generation_df.sort_values("date").reset_index(drop=True)
```

```python
print(generation_df["date"].min())
print(generation_df["date"].max())
```

```python
consumption_df.to_csv(os.path.join(processed_dir, "consumption.csv"), index=False)
generation_df.to_csv(os.path.join(processed_dir, "generation.csv"), index=False)
```

```python
consumption_df.to_parquet(os.path.join(processed_dir, "consumption.parquet"))
generation_df.to_parquet(os.path.join(processed_dir, "generation.parquet"))
```
